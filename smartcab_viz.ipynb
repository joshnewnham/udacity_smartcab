```python
cmap=plt.cm.Oranges
import numpy as np
def plot_matrix(m, title='Q-Matrix weights', action_labels=None, state_labels=None):
    
    state_labels = []
    q_values = [] 
    for state, q_value in m.iteritems():
        state_labels.append(state)
        q_values.append(q_value)
    
    small_im = np.array([[0.25, 0.75, 1.0, 0.75], [0.1, 0.65, 0.5, 0.4],
                     [0.6, 0.3, 0.0, 0.2], [0.7, 0.9, 0.4, 0.6]])
    
    plt.imshow(q_values, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    x_tick_marks = np.arange(len(action_labels))
    y_tick_marks = np.arange(len(state_labels))
    plt.xticks(x_tick_marks, action_labels, rotation=45)
    plt.yticks(y_tick_marks, state_labels)
    plt.ylabel('States')
    plt.xlabel('Actions')
    plt.show()
```
